# Libraries

```python
#General purpose
import pandas as pd

#Hugging face
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline
```

# Instances

```python
# Emotional analysis

emotional_tokenizer = AutoTokenizer.from_pretrained("pysentimiento/robertuito-emotion-analysis")

emotional_model = AutoModelForSequenceClassification.from_pretrained("pysentimiento/robertuito-emotion-analysis")
```

```python
# Sentimental analysis

sent_tokenizer = AutoTokenizer.from_pretrained("pysentimiento/robertuito-sentiment-analysis")

sent_model = AutoModelForSequenceClassification.from_pretrained("pysentimiento/robertuito-sentiment-analysis")
```

```python
# Hate speech

hate_tokenizer = AutoTokenizer.from_pretrained("pysentimiento/robertuito-hate-speech")

hate_model = AutoModelForSequenceClassification.from_pretrained("pysentimiento/robertuito-hate-speech")
```

# Try

```python
text = "No estoy de acuerdo con hacer mal las cosas"
```

## Emotional Analysis

```python
inputs = emotional_tokenizer(text, return_tensors="pt")

text_to_emotion = pipeline('sentiment-analysis', model='pysentimiento/robertuito-emotion-analysis')

result = text_to_emotion(text)
print(result)
```

## Sentimental analysis

```python
inputs = sent_tokenizer(text, return_tensors="pt")

text_to_sent = pipeline('sentiment-analysis', model='pysentimiento/robertuito-sentiment-analysis')

result = text_to_sent(text)
print(result)
```

## Hate analysis

```python
inputs = hate_tokenizer(text, return_tensors="pt")

text_to_hate = pipeline('sentiment-analysis', model='pysentimiento/robertuito-hate-speech')

result = text_to_hate(text)
print(result)
```
